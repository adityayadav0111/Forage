```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Task 3_Final Content Data set.csv',index_col=0)
```

```python
df
```

```python
df.groupby('Category').sum('Score').head().plot(kind='bar')
plt.savefig('Top_category_based_on_score.png') 
```

```python
df['Category'].value_counts().plot(kind='pie',autopct='%0.2f%%')
plt.savefig('pie_with_all_categories .png') 
```
